AI Assistant Chatbot Development


```python
import ollama
```

```python
response=ollama.chat(
    model='mistral',
    messages=[
        {
            'role':'user',
            'content':'what is deep learning?'
        }
    ]
)
print(response['message']['content'])
```

```python
user_input=input("Ask question...")
response=ollama.chat(
    model='mistral',
    messages=[
        {
            'role':'user',
            'content': user_input
        }
    ]
)
print(response['message']['content'])
```

```python
def ask_ai(question):
    response = ollama.chat(
        model='mistral',
        messages=[
            {
                'role': 'user',
                'content': question
            }
        ]
    )
    return response['message']['content']
```

```python
def gym_trainer(question):
    response = ollama.chat(
        model='mistral',
        messages=[
            {
                'role': 'user',
                'content': """You are a gym trainer.
                rules:
                - be polite and friendly
                - keep answers short and to the point
                - give diet plan for weight loss
                """
            },
            {
                'role': 'user',
                'content': question
            }
        ]
    )
    return response['message']['content']
```

```python
gym_trainer("i want a diet plan to lose weight")
```
